```python
%matplotlib inline
import numpy as np      
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

from skimage import measure
import scipy.ndimage as ndi
from pylab import rcParams
rcParams['figure.figsize'] = (6, 6)
```

#### 原图像

```python
# 查看图片
img = mpimg.imread('../rawdata/images/218.jpg')
plt.imshow(img)
plt.show()
```

#### 提取轮廓

```python
contours = measure.find_contours(img, .8)
contour = max(contours, key=len)

f = plt.figure(figsize=(5,5))
plt.plot(contour[::,-1], max(contour[::,0])+min(contour[::,0])-contour[::,0], linewidth=1)
plt.show()
```

```python
def cart2pol(x, y):
    """平面直角坐标系转换为极坐标系"""
    rho = np.sqrt(x**2 + y**2)
    phi = np.arctan2(y, x)
    return [rho, phi]

# 以重心为中心转换为极坐标
cy, cx = ndi.center_of_mass(img)
contour[::,1] -= cx
contour[::,0] -= cy

polar_contour = np.array([cart2pol(x, y) for x, y in contour])
f = plt.figure(figsize=(16,6))
plt.scatter(polar_contour[::,1], polar_contour[::,0], 1)
plt.show()
```

```python
# 仅考虑长度rho
f = plt.figure(figsize=(16,6))
plt.scatter(range(len(polar_contour[::,0])),polar_contour[::,0],1)
```

```python
# 仅考虑弧度phi
f = plt.figure(figsize=(16,6))
plt.scatter(range(len(polar_contour[::,1])),polar_contour[::,1],1)
```

#### 对每张图分别提取形状和边缘特征，并保存

```python
from tqdm import tqdm_notebook as tqdm

shape_contour_list = []
blade_contour_list = []
for id_ in tqdm(range(1,1585)):
    tmp_features = extract_features(id_)
    shape_contour_list.append(tmp_features['shape_series'][0])
    blade_contour_list.append(tmp_features['blade_series'][0])

shape_contour_array = np.array(shape_contour_list)
blade_contour_array = np.array(blade_contour_list)
# 保存
np.save('../middata/shape_contour_array.npy',shape_contour_array)
np.save('../middata/blade_contour_array.npy',blade_contour_array)
```
